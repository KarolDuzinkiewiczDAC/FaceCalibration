# Notebook to illustrate 'naive' approach to camera calibration based on 3D face landmarks

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# load example data from phone
filepath = '../wet/data4/frame_kpts_1.npy'

with open(filepath, 'rb') as f:
    frame = np.load(f)
    face_keypoints = np.load(f)

assert frame is not None
assert len(face_keypoints) == 478
```

```python
# display Face Blaze landmarks
img = frame.copy()
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
for i, landmark in enumerate(face_keypoints):
    img = cv2.drawMarker(img, (int(landmark[0]), int(landmark[1])), color=(255, 0, 0))

plt.imshow(img)
plt.show()
```
